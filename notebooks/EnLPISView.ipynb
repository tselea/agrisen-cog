# English Version of LPIS View

This Jupyter notebook is designed to read Land Parcel Identification System (LPIS) **pre-processed English version** aois = ['Austria', 'Belgium', 'Catalonia','Denmark', 'Netherlands']
years = [2019,2020]files for various areas of interest (AOIs) across different years. 

```python
import dask_geopandas
from IPython.display import display, HTML
```

```python
aois = ['Austria', 'Belgium', 'Catalonia','Denmark', 'Netherlands']
years = [2019,2020]
```

## S3 bucket Partioned Parquet files

```python
for aoi in aois:
    for year in years:
        file_name = f's3://agrisen-cog-v1/LPIS_processing/en_files/{aoi}_{year}_en_distrib.parquet'
        df_dgpd = dask_geopandas.read_parquet(file_name,storage_options={
            "anon": True,
                    "client_kwargs": {"endpoint_url": 'https://s3-3.services.tselea.info.uvt.ro'}
                }) 
        print(f'{aoi} {year} has {df_dgpd.shape[0].compute()} rows.')
```

```python
for aoi in aois:
    for year in years:
        file_name = f's3://agrisen-cog-v1/LPIS_processing/en_files/{aoi}_{year}_en_distrib.parquet'
        df_dgpd = dask_geopandas.read_parquet(file_name,storage_options={
            "anon": True,
                    "client_kwargs": {"endpoint_url": 'https://s3-3.services.tselea.info.uvt.ro'}
                }) 
        print(f'{aoi} {year}')
        display(HTML(df_dgpd.head().to_html()))
```
